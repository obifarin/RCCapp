This script was used in pre-processing the data for the RCC metabolomics web app.

```python
import pandas as pd
import numpy as np
```

```python
np.__version__
```

```python
 # Import the excel file containing all the combined NMR and MS chemical data, with a sample ID for each patient.
RCCdf = pd.read_excel('data/RCCdataframe.xlsx')
    # Import the excel file that contains the sample ID (chemistry), the patient ID (clinical) for each patient
    # inaddition to some relevant meta-data/
MSrcclink = pd.read_excel('data/Sample_List_MS.xlsx')
    
    # select specific biomarker and only patient ID.
    # AND #
    # Metadata to potentially add: 'Nuclear Grade', 'Grade', 'Stage', 'Grouped Stage',
biomarker = RCCdf[['Sample ID', 6954, 6578, 6262, 3804, 2102, 1481, 720, 'Metastatic', 'SubTypes', 'Stage']]

    # copy the data
biomarker_scaled = biomarker.copy()

    # Select all metabolite column
column = [6954, 6578, 6262, 3804, 2102, 1481, 720]
    # Log transform metabolite abundance.
biomarker_scaled[column] = np.log(biomarker_scaled[column])

    # MSrcclink dataframe that contains the sample ID (chemistry), and
    # the patient ID (clinical) for each patient In addition to some relevant meta-data/
final_df = biomarker_scaled.merge(MSrcclink.iloc[:, :7])
final_df.rename(columns={6954: "N-acetyl-glucosaminic acid",
                         6578: "2-mercaptobenzothiazole",
                         6262: "hippurate-mannitol derivative",
                         3804: "hippuric acid",
                         2102: "dibutylamine",
                         1481: "Lys-Ile",
                         720: "2-phenylacetamide"}, inplace=True)

    #rename Stage columns entries
dmap = {1: "I", 2: "II", 3: "III", 4:"IV"}
final_df["Stage"] = final_df["Stage"].map(dmap)
```
